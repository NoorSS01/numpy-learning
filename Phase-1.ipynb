## Numpy array and basics

```python
import numpy as np
```

creating array from List

```python
arr1d=np.array([1,2,3])
print(f"1D Array : {arr1d}")

arr2d=np.array([[1,2,3],[1,2,3]])
print(f"2D Array : {arr2d}")
```

### List vs Numpy Array

```python
py_list=[1,2,3]
print(f"Python list Multiplication :{py_list * 2}")

np_array=np.array([1,2,3])
print(f"Python array Multiplication : {np_array * 2}")

import time
start=time.time()
py_list=[i*2 for i in range(1000000)]
print("\n List operation time : ", time.time()-start)

start=time.time()
np_array=np.array(10000000)*2
print("\n Numpy operation time:", time.time() - start)
```

### creating arrays from scratch

```python
zeros=np.zeros((3,4))
print(f" zeros array : {zeros}")

ones=np.ones((2,3))
print(f" ones array : {ones}")

full=np.full((4,4), 7)
print(f"full array :{full}")

random=np.random.random((3,4))
print(f"random array : {random}")

sequence=np.arange(0,10,2)
print(f"random array : {sequence}")
```

### Vector Matrix and Tensor

```python
vector=np.array([1,2,3])
print(f"Vector : {vector}")

matrix=np.array([[1,2,3],
                 [1,2,4]])
print(f"Matrix : {matrix}")

tensor=np.array([[[1,2],[3,4]],
                 [[5,6],[7,8]]])
print(f"Tensor : {tensor}")
```

### Array Properties

```python
arr=np.array([[1,2,3],
              [4,5,6]])
print(f"Shape : {arr.shape}")
print(f"Dimension : {arr.ndim}")
print(f"Size : {arr.size}")
print(f"DType : {arr.dtype}")
```

```python
arr=np.arange(12)
print(f"Original array {arr}")

reshaped=arr.reshape((3,4))
print(f"Reshaped array : {reshaped}")

flattened= reshaped.flatten()
print(f"Flattened array : {flattened}")

#Raveled (returns view, instead of copy )
raveled=reshaped.ravel()
print(f"Raveled array : {raveled}")

#Transpose
Transpose=reshaped.T
print(f"Transpose array {Transpose}")
```
